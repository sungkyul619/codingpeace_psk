```python
# 바운딩박스 리스트를 욜로포멧팅 문자열로 전환하는 함수
def format(jason_data):
  labeling = jason_data['annotations'][0]['bbox']
  prelabeling = str(labeling[0])+' '+str(labeling[1])
  str_list = ((prelabeling.replace('[','')).replace(']','')).replace(',','')
  return str_list
```

```python
# 리스케일 진행하기
from PIL import Image

def convert(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return [x,y,w,h]
  
```

```python
newlabels_path=[]
for index,value in enumerate(jpg_list[2]):
  img_path = f'{com_path}test/images/{value}'
  im=Image.open(img_path)
  w= int(im.size[0])
  h= int(im.size[1])

  newlabels=value.replace('jpg','json.txt')
  newlabels_path= f'{com_path}test/newlabels/{newlabels}'
 
 # 읽기
  f = open(newlabels_path, 'r', encoding='utf-8')
  lines = f.readlines()

 # print('파일 읽기',lines) # lines는 list타입
  strlabel = lines[0].split()

  xmin= float(strlabel[0])
  ymin=float(strlabel[1])
  xmax =float(strlabel[2])
  ymax =float(strlabel[3])
 
  #print('xmin, xmax, ymin, ymax',xmin, xmax, ymin, ymax) #define your x,y coordinates
  b = (xmin, xmax, ymin, ymax)
  bb = convert((w,h), b) 
  #print('리스케일링',bb) # bb 는 list, 크기는 4
  new_value = f"{bb[0]} {bb[1]} {bb[2]} {bb[3]}"
  print(new_value)

  f.close()

  # # 쓰기
  f = open(newlabels_path, 'w')
  f.write(new_value)
  f.close()
```
